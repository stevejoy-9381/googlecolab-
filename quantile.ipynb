```python
import pandas  as pd
import numpy as np
```

```python
df=pd.read_csv('/content/heights (2).csv')
df.head(20)
```

```python
df.describe()
```

```python
q1=df.height.quantile(0.25)
q1
```

```python
q3=df.height.quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower_limit=q1-1.5*iqr
lower_limit
```

```python
upper_limit=q3+1.5*iqr
upper_limit
```

```python
df[(df.height<lower_limit) | (df.height>upper_limit)]
```

```python
df2=df[(df.height<upper_limit) & (df.height >lower_limit)]
df2.head()
```
