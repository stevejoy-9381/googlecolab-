```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/personality_dataset.csv')
```

```python
df1=df.dropna()
```

```python
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
stage_fear_n=LabelEncoder()
Drained_after_socializing_n=LabelEncoder()
Personality_n=LabelEncoder()
```

```python
df1['stage_fear_n']  = stage_fear_n.fit_transform(df1['Stage_fear'])
df1['Drained_after_socializing_n']  = Drained_after_socializing_n.fit_transform(df1['Drained_after_socializing'])
df1['Personality'] = Personality_n.fit_transform(df1['Personality'])
```

```python
df1.head(10)
```

```python
df3=df1.drop(['Stage_fear','Drained_after_socializing'],axis='columns')
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
x=df3.drop(['Personality'],axis='columns')
```

```python
y=df3.Personality
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[9.0,1,1.0	,2.0,	1,	5.0	,2.0]])
```
