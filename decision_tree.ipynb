```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/salaries.csv')
df
```

```python
input_drop=df.drop('salary_more_then_100k',axis='columns')
```

```python
input_drop
```

```python
target=df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
n_company=LabelEncoder()
n_job=LabelEncoder()
n_degree=LabelEncoder()
```

```python
input_drop['company_n']=n_company.fit_transform(input_drop['company'])
input_drop['job_n']=n_job.fit_transform(input_drop['job'])
input_drop['degree_n']=n_degree.fit_transform(input_drop['degree'])
input_drop
```

```python
input_drop=input_drop.drop(['company','job','degree'],axis='columns')
input_drop
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(input_drop,target,test_size=0.1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
