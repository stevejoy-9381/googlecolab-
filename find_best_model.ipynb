```python
import pandas as pd
from sklearn.datasets import load_iris
flow=load_iris()
```

```python
df=pd.DataFrame(flow.data,columns=flow.feature_names)
df.head()
```

```python
df
```

```python
df['target']=flow.target
df['target']=df['target'].apply(lambda x:flow.target_names[x])
df.head()
```

```python
x=df.drop(['target'],axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn import svm
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),x,y,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),x,y, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),x,y, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='linear',C=20,gamma='auto'),x,y,cv=5)
```

```python
import numpy as np
```

```python
kernal=['rbf,linear']
cv=[10,20,30]
avg_scores={}
for kval in kernal:
  for cva in cv:
    cv_scores=cross_val_score(svm.SVC(kernel=kval,C=cva,gamma='auto'),x,y,cv=5)
    avg_scores[kval + '_' + str(cva)] = np.average(cv_scores)
avg_scores
```

```python
kernels = ['rbf', 'linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),x,y, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)

avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, return_train_score=False)
clf.fit(x,y)
result=clf.cv_results_
result
```

```python
df=pd.DataFrame(result)
df
```

```python
df[['param_C','params','mean_test_score']]
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python

from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'), {
        'C': [1,10,20],
        'kernel': ['rbf','linear']
    },
    cv=5,
    return_train_score=False,
    n_iter=2
)
rs.fit(x,y)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```
